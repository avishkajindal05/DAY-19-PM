Simulate and compare distributions:
1. Generate samples from:
○ Normal distribution
○ Uniform distribution
2. Plot both distributions and compare:
○ Shape
○ Spread
○ Central tendency
3. Explain:
○ When would you use each distribution in real-world problems?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
normal = np.random.normal(0,1,1000)
uniform = np.random.uniform(-2,2,1000)

plt.hist(normal, alpha=0.5, bins=30)
plt.hist(uniform, alpha=0.5, bins=30)
plt.legend(["Normal", "Uniform"])
plt.show()
```

Explanation:
Normal → bell-shaped
Uniform → flat
Use:
Normal → natural phenomena
Uniform → random sampling
